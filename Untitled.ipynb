```python
import numpy as np
import matplotlib.pyplot as plt

# Provided datasets
exact_data = {
    "4": [0.000621, 0.000571, 0.00054, 0.000637],
    "9": [0.016638, 0.019247, 0.019413, 0.017618],
    "16": [0.545504, 0.543159, 0.566653, 0.549391],
    "25": [304.617491, 357.131427, 360.643962, 392.051544],
}

qaoa_data = {
    '4': [0.40901, 0.521832, 0.374727, 0.538126], 
    '9': [1.734273, 1.596064, 2.531626, 1.935299], 
    '16': [23.204654, 27.400321, 21.675526, 21.877055],
    "25" :[830.258702, 953.383286]
}

qa_data = {
    '4': [0.007508, 0.011068, 0.010892, 0.014196], 
    '9': [0.01009, 0.008866, 0.011314, 0.008578], 
    '16': [0.0010364, 0.0010076, 0.010752, 0.01034], 
    '25': [0.0011096, 0.0014168, 0.009706, 0.012246],
    '400': [20.996/100]
}

# Function to calculate averages
def calculate_averages(data):
    N_values = sorted(int(k) for k in data.keys())
    averages = [np.mean(data[str(N)]) for N in N_values]
    return N_values, averages

# Calculate averages for each dataset
N_exact, avg_exact = calculate_averages(exact_data)
N_qaoa, avg_qaoa = calculate_averages(qaoa_data)
N_qa, avg_qa = calculate_averages(qa_data)

# Initialize the plot
plt.figure(figsize=(10, 6))

# Define colors for each dataset
colors = {
    'Exact': 'blue',
    'QAOA': 'green',
    'QA': 'red'
}

# Plot each dataset's averages
plt.scatter(N_exact, avg_exact, color=colors['Exact'], label='Exact Data', alpha=0.7)
plt.scatter(N_qaoa, avg_qaoa, color=colors['QAOA'], label='QAOA Data', alpha=0.7)
plt.scatter(N_qa, avg_qa, color=colors['QA'], label='QA Data', alpha=0.7)

# Function to add trend lines
def add_trend_line(N, averages, color, label):
    # Fit a linear trend line
    z = np.polyfit(N, averages, 1)
    p = np.poly1d(z)
    plt.plot(N, p(N), linestyle='--', color=color, label=f'{label} Trend Line')

# Plot each dataset's averages and connect the dots
plt.plot(N_exact, avg_exact, color=colors['Exact'],  alpha=0.7, linestyle='-', marker='o')
plt.plot(N_qaoa, avg_qaoa, color=colors['QAOA'], alpha=0.7, linestyle='-', marker='o')
plt.plot(N_qa, avg_qa, color=colors['QA'], alpha=0.7, linestyle='-', marker='o')


# Adding plot details
plt.xlabel("Lattice Size (# Spins)", fontsize=12)
plt.ylabel("Computation Time (sec)", fontsize=12)
plt.yscale('log')  # Logarithmic scale for better visualization
plt.legend()
#plt.grid(True, which="both", linestyle="--", linewidth=0.5, alpha=0.7)
plt.tight_layout()

# Display the plot
plt.rcParams['font.family'] = 'serif'  # Options: 'serif', 'sans-serif', 'monospace', etc.

plt.savefig("plot_high_res.png", dpi=300)  # High-resolution PNG
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Provided datasets
exact_datas= {
    "4": [100, 100, 100, 100],
    "9": [100, 100, 100, 100],
    "16": [100, 100, 100, 100],
    "25": [100, 100, 100, 100]}
    


qaoa_datas= {
    '4': [100], 
    '9': [84.96], 
    '16': [63.56],
}

qa_datas = {
    "4": [100,100, 100, 100],
    "9": [100, 100, 100, 100],
    "16": [100, 100, 100, 100],
    "25": [100, 100, 100, 100],
}


# Function to calculate averages
def calculate_averages(data):
    N_values = sorted(int(k) for k in data.keys())
    averages = [np.mean(data[str(N)]) for N in N_values]
    return N_values, averages

# Calculate averages for each dataset
N_exact, avg_exact = calculate_averages(exact_datas)
N_qaoa, avg_qaoa = calculate_averages(qaoa_datas)
N_qa, avg_qa = calculate_averages(qa_datas)

# Initialize the plot
plt.figure(figsize=(10, 6))

# Define colors for each dataset
colors = {
    'Exact': 'blue',
    'QAOA': 'green',
    'QA': 'red'
}

# Define line styles for differentiation
line_styles = {
    'Exact': '-',
    'QAOA': '--',
    'QA': ':'
}

# Plot each dataset's averages
plt.scatter(N_exact, avg_exact, color=colors['Exact'],  alpha=0.7, marker='o')
plt.scatter(N_qaoa, avg_qaoa, color=colors['QAOA'],  alpha=0.7, marker='s')
plt.scatter(N_qa, avg_qa, color=colors['QA'],  alpha=0.7, marker='^')

# Plot trend lines with distinct line styles
plt.plot(N_exact, avg_exact, color=colors['Exact'], linestyle=line_styles['Exact'], marker='o', label='Exact Data')
plt.plot(N_qaoa, avg_qaoa, color=colors['QAOA'], linestyle=line_styles['QAOA'], marker='s', label='QAOA Data')
plt.plot(N_qa, avg_qa, color=colors['QA'], linestyle=line_styles['QA'], marker='^', label='QA Data')

# Adding plot details
plt.xlabel("Lattice Size (# Spins)", fontsize=12)
plt.ylabel("Computational Accuracy (%)", fontsize=12)
plt.legend()
plt.tight_layout()

# Set font family and save the plot
plt.rcParams['font.family'] = 'serif'  # Options: 'serif', 'sans-serif', 'monospace', etc.
plt.savefig("plot_high_res_distinct_styles.png", dpi=300)  # High-resolution PNG
plt.show()

```
