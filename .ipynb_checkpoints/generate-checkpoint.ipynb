```python
import numpy as np

def generate_h(n, m, save=True):
    h = np.random.rand(n, m) * 2 - 1
    if save:
       np.save(f'h_{n}_{m}.npy', h) 
    return h
```

```python
h = generate_h(5, 5)
print(h)
```

```python
h = generate_h(10, 10)
print(h)
```

```python
h = generate_h(6, 6)
print(h)
```

```python
h = generate_h(30, 30)
print(h)
```
